```python
import os
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient

# Substitua pelos valores reais obtidos no Passo 1
account_name = 'uvabr'
account_key = os.environ['ACCOUNT_KEY']

# Conecte-se ao servi√ßo de armazenamento de blob
connect_str = f"DefaultEndpointsProtocol=https;AccountName={account_name};AccountKey={account_key};EndpointSuffix=core.windows.net"
blob_service_client = BlobServiceClient.from_connection_string(connect_str)

# Fun√ß√£o para baixar arquivos de um cont√™iner
def download_blobs_from_container(container_name, download_path):
    container_client = blob_service_client.get_container_client(container_name)
    
    # Cria o diret√≥rio de download se n√£o existir
    if not os.path.exists(download_path):
        os.makedirs(download_path)
    
    # Itera sobre todos os blobs no cont√™iner
    blob_list = container_client.list_blobs()
    for blob in blob_list:
        blob_client = container_client.get_blob_client(blob)
        download_file_path = os.path.join(download_path, blob.name)
        
        # Cria o diret√≥rio local se n√£o existir
        os.makedirs(os.path.dirname(download_file_path), exist_ok=True)
        
        print(f"Baixando {blob.name} para {download_file_path}")
        with open(download_file_path, "wb") as download_file:
            download_file.write(blob_client.download_blob().readall())

# Diret√≥rios locais de download
download_dirs = {
    'bronze': 'path/to/local/bronze',
    'silver': 'path/to/local/silver',
    'gold': 'path/to/local/gold'
}

# Baixa os arquivos de cada cont√™iner
for container_name, download_path in download_dirs.items():
    download_blobs_from_container(container_name, download_path)

print("Download conclu√≠do.")
```
